# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    return x * (x > 0)
def dReLU(x):
    return 1 * (x > 0)

x = plt.linspace(-10, 10, 100)

f1, = plt.plot(x, ReLU(x), 'r')
f2, = plt.plot(x, dReLU(x), 'b')

plt.grid()

plt.title('ReLU Function')

plt.legend([f1, f2], ['ReLU', 'Derivative of ReLU'], loc='upper left', fontsize=12)

plt.show()
```
